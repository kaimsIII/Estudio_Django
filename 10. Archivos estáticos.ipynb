## 📌 **Archivos estáticos en Django (CSS, JS, imágenes)**  

El objetivo de este tema es que comprendas cómo Django maneja archivos estáticos (CSS, JavaScript, imágenes, fuentes, etc.), y cómo usarlos en tus proyectos.  

---

### 🔹 **1. ¿Qué son los archivos estáticos y por qué son importantes?**  

Los **archivos estáticos** son aquellos que no cambian con cada solicitud del usuario. Incluyen:  

- **CSS** → Para los estilos de la aplicación.  
- **JavaScript** → Para funcionalidades interactivas en el frontend.  
- **Imágenes** → Para iconos, logos, fondos, etc.  
- **Fuentes y otros archivos** → Como archivos de fuentes personalizadas.  

Django **no sirve automáticamente los archivos estáticos en producción**, por lo que debemos configurarlos correctamente.  

---

### 🔹 **2. Configuración de archivos estáticos en Django**  

📍 **Paso 1: Definir la carpeta de archivos estáticos**  

Django ya tiene una configuración por defecto en `settings.py`:  

```python
import os

STATIC_URL = '/static/'
STATICFILES_DIRS = [
    os.path.join(BASE_DIR, "static"),  # Carpeta donde colocaremos los archivos
]
```

📌 **Explicación:**  
✔️ `STATIC_URL` define la URL base para archivos estáticos (`/static/`).  
✔️ `STATICFILES_DIRS` le dice a Django dónde buscar archivos estáticos en modo desarrollo.  

📍 **Paso 2: Crear la carpeta `static/` en el proyecto**  

Dentro de la estructura del proyecto, podemos crear una carpeta `static/` para almacenar archivos:  

```
mi_proyecto/
│── mi_app/
│── static/
│   ├── css/
│   │   ├── estilos.css
│   ├── js/
│   │   ├── script.js
│   ├── img/
│   │   ├── logo.png
│── templates/
│── settings.py
```

---

### 🔹 **3. Usando archivos estáticos en plantillas HTML**  

Para que Django reconozca los archivos estáticos en las plantillas, usamos `{% load static %}`.  

📍 **Ejemplo de plantilla (`base.html`) con archivos estáticos:**  

```html
{% load static %}

<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Mi Proyecto</title>
    <link rel="stylesheet" href="{% static 'css/estilos.css' %}">
</head>
<body>

    <h1>Bienvenido a mi sitio</h1>
    <img src="{% static 'img/logo.png' %}" alt="Logo">
    <script src="{% static 'js/script.js' %}"></script>

</body>
</html>
```

📌 **Explicación:**  
✔️ `{% load static %}` carga el sistema de archivos estáticos de Django.  
✔️ `{% static 'ruta/archivo' %}` genera la URL correcta para acceder al archivo.  

---

### 🔹 **4. Uso avanzado: Archivos estáticos dentro de una app**  

En proyectos más organizados, cada **app** dentro de Django puede tener su propia carpeta `static/`.  

📍 **Estructura con archivos estáticos dentro de una app**  

```
mi_proyecto/
│── mi_app/
│   ├── static/
│   │   ├── mi_app/
│   │   │   ├── css/
│   │   │   │   ├── estilos.css
│   │   │   ├── js/
│   │   │   │   ├── script.js
│   │   │   ├── img/
│   │   │   │   ├── logo.png
│── templates/
│── settings.py
```

Para referenciar estos archivos en plantillas:  

```html
<link rel="stylesheet" href="{% static 'mi_app/css/estilos.css' %}">
```

Django buscará dentro de `mi_app/static/mi_app/css/estilos.css`.  

---

### 🔹 **5. Sirviendo archivos estáticos en producción**  

En producción, los archivos estáticos deben estar en un solo directorio accesible por el servidor web. Para esto, agregamos en `settings.py`:  

```python
STATIC_ROOT = os.path.join(BASE_DIR, "staticfiles")
```

Luego, ejecutamos:  

```bash
python manage.py collectstatic
```

Esto recopila todos los archivos estáticos en `STATIC_ROOT`, listos para ser servidos en producción.  

---

### ✅ **Resumen del flujo de trabajo con archivos estáticos**  

1️⃣ Definimos la configuración en `settings.py` (`STATIC_URL` y `STATICFILES_DIRS`).  
2️⃣ Creamos la carpeta `static/` y organizamos archivos CSS, JS, imágenes, etc.  
3️⃣ Usamos `{% load static %}` y `{% static 'archivo' %}` en las plantillas.  
4️⃣ En producción, ejecutamos `collectstatic` para recopilar los archivos en `STATIC_ROOT`.
