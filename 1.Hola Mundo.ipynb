### **3. Estructura de un proyecto Django**  

Cuando creamos un proyecto, Django genera una serie de carpetas y archivos esenciales para su funcionamiento. Antes de escribir c√≥digo, es importante entender la l√≥gica detr√°s de esta estructura.

### **3.1. ¬øQu√© es un proyecto en Django?**  
Un **proyecto** en Django es la configuraci√≥n general de una aplicaci√≥n web. Contiene archivos que manejan:  
‚úÖ Configuraci√≥n del servidor web  
‚úÖ Conexi√≥n con bases de datos

üîπ **Un proyecto puede contener varias aplicaciones**, cada una encargada de una funcionalidad espec√≠fica.

### **3.2. Creando un proyecto Django**  
Para generar un nuevo proyecto, usamos el comando:  

```bash
django-admin startproject mi_proyecto
```
üìå **Explicaci√≥n:**  
- `django-admin` ‚Üí Es una herramienta de Django para ejecutar comandos.  
- `startproject` ‚Üí Crea una nueva estructura base de un proyecto.  
- `mi_proyecto` ‚Üí Es el nombre del proyecto (puedes cambiarlo).  

Despu√©s de ejecutar el comando, se crear√° la siguiente estructura:  

```
mi_proyecto/
‚îÇ-- manage.py
‚îÇ-- mi_proyecto/
‚îÇ   ‚îÇ-- __init__.py
‚îÇ   ‚îÇ-- settings.py
‚îÇ   ‚îÇ-- urls.py
‚îÇ   ‚îÇ-- asgi.py
‚îÇ   ‚îÇ-- wsgi.py
```

### **3.3. Explicaci√≥n de los archivos y carpetas**  

#### üìÇ **Carpeta del proyecto (`mi_proyecto/`)**  
Aqu√≠ es donde se almacenan los archivos principales del proyecto.

üîπ **Archivos clave dentro de esta carpeta:**  

| Archivo       | Explicaci√≥n |
|--------------|------------|
| `__init__.py` | Indica que esta carpeta es un m√≥dulo de Python. |
| `settings.py` | Archivo de configuraci√≥n global del proyecto (base de datos, seguridad, apps, etc.). |
| `urls.py` | Define las rutas del proyecto (c√≥mo se accede a las vistas). |
| `asgi.py` | Permite ejecutar Django con servidores ASGI (para aplicaciones en tiempo real). |
| `wsgi.py` | Permite ejecutar Django con servidores WSGI (para producci√≥n). |

#### üìÑ **`manage.py`**  
Este es un script clave para interactuar con el proyecto. Nos permite ejecutar comandos como:  
- **Levantar el servidor:** `python manage.py runserver`  
- **Crear una aplicaci√≥n:** `python manage.py startapp mi_app`

### **4. Crear una aplicaci√≥n en Django**  
Django divide los proyectos en **aplicaciones**. Mientras que un **proyecto** es la configuraci√≥n general, una **aplicaci√≥n** representa una funcionalidad espec√≠fica.  

Por ejemplo:

Para crear una aplicaci√≥n, usamos:

üìå **Explicaci√≥n de los archivos dentro de una aplicaci√≥n:**  

| Archivo       | Explicaci√≥n |
|--------------|------------|
| `migrations/` | Archivos de migraci√≥n de la base de datos. |
| `__init__.py` | Indica que esta carpeta es un m√≥dulo de Python. |
| `admin.py` | Configuraci√≥n del panel de administraci√≥n de Django. |

### **4.1. Registrar la aplicaci√≥n en el proyecto**  
Para que Django reconozca la aplicaci√≥n, debemos registrarla en `settings.py`.

Ahora Django reconocer√° la aplicaci√≥n y podremos usarla en nuestro proyecto.

### **5.1. Creando una vista**  
Una **vista** es una funci√≥n que maneja una solicitud web y devuelve una respuesta (puede ser texto, HTML, JSON, etc.).  

En **`views.py`** de la aplicaci√≥n `mi_app`, agrega lo siguiente:

üìå **Explicaci√≥n:**  
- `HttpResponse` ‚Üí Permite devolver texto en la respuesta.

### **5.2. Agregando la vista a las URLs**  
Para acceder a esta vista desde el navegador, debemos definir una URL.  

Crea un archivo `urls.py` dentro de `mi_app` y agrega:  

```python
from django.urls import path
from . import views

urlpatterns = [
    path('hola/', views.hola_mundo),
]
```

```python
from django.contrib import admin
from django.urls import path, include

urlpatterns = [
    path('admin/', admin.site.urls),
    path('mi_app/', include('mi_app.urls')),  # Incluir URLs de la aplicaci√≥n
]
```

üìå **Explicaci√≥n:**  
- `include('mi_app.urls')` ‚Üí Importa las rutas de `mi_app`.  
- `path('mi_app/', ...)` ‚Üí Todas las URLs de `mi_app` estar√°n dentro del prefijo `mi_app/`.
