### **1.3. Registrar Modelos en el Admin**  
Por defecto, el panel est√° vac√≠o. Para gestionar nuestros modelos (`Usuario`), debemos registrarlos en `admin.py`:

### **1.4. Personalizar la Vista del Admin**  
Podemos mejorar la presentaci√≥n del modelo en el panel de administraci√≥n.  

üìå **Modifica `admin.py` para personalizar la vista:**  
```python
class UsuarioAdmin(admin.ModelAdmin):
    list_display = ('id', 'nombre', 'edad', 'email')  # Columnas en la lista
    search_fields = ('nombre', 'email')  # Agregar barra de b√∫squeda
    list_filter = ('edad',)  # Filtros en el panel
    ordering = ('nombre',)  # Ordenar por nombre

admin.site.register(Usuario, UsuarioAdmin)
```
